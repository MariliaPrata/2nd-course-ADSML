###  Convert Calories to Kilojoules

```python
# Define function to convert calories to kilojoules

def caltokj(x):
    
    return x*4.184   
```

###  Apply filter across the entire Dataframe

Apply this step to all the five columns at once. If we compute the filter across the entire DataFrame we just obtain a DataFrame of boolean values.

```python
#Write down a boolean condition. Looking at the absolute value of df3 - df3.mean(axis=0)
#See if it is greater than 3 * df3.std(axis=0). Use numpy’s function abs() to take the absolute value


zfilter_df = np.abs(df3 - df3.mean(axis=0)) > (3 * df3.std(axis=0))
```

### Try boxplot method

Identify the outliers flagged in the boxplot method by calculating the upper and lower bounds directly from the IQR of each feature.

```python
#I tried to fillna with the mean() cause I got Nan on Lower and upper bound

df3.fillna(df3.mean(), inplace=True)
```

```python
#Quartiles and the interquartile range are robust statistics that are unaffected 
#by this small set of data points.

Q1 = np.percentile(df3, 25, axis=0)
Q3 = np.percentile(df3, 75, axis=0)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Print lower and upper bound for X0
print("Lower bound: {:.4f}, upper bound: {:.4f}".format(lower_bound[0], upper_bound[0]))
```

```python
#Use boolean indexing to select the entries which lie outside our bounds. 
#Check how many outliers are identified by the boxplot in each feature.

boxplot_filter = (df3 < lower_bound) | (df3 > upper_bound)
print("Number of outliers detected per feature using a boxplot:")
print(boxplot_filter.sum(axis=0))
print("Total number of outliers detected:", boxplot_filter.sum(axis=0).sum())
```
